```python
import urllib.request as req
url = "https://github.com/Elwing-Chou/kmu1202/raw/main/danmu.txt"
req.urlretrieve(url, "danmu.txt")
```

```python
import pandas as pd
# pandas自訂型態: 1. DataFrame: 表格  2. Series: 一列/一行
df = pd.read_json("danmu.txt", encoding="utf-8")
df
```

```python
# Demo 篩選
# df[跟你的資料筆數依樣多的True/False]
test = pd.DataFrame([
    [1, 2],
    [3, 4],
    [5, 6]
])
fil = [True, False, False]
test[fil]
```

```python
fil = df["time"] > 10000
df[fil]
```

```python
# 而且(&) 或者(|)
fil = (df["time"] > 10000) & (df["time"] < 30000) & (df["color"] == "#FFFFFF")
df[fil]
```

```python
# 轉換操作
print(df["time"] * 2 + 1)
print(df["color"] + "abc")
```

```python
# Everything is an object
# 每個東西都是具有型態
# 型態都具有它特定的操作
# 字串型態: s = "hello", 操作: s[0]
# 字典型態: d = {"height":175} 操作: d["height"]
# 還沒執行的SOP型態: int 操作:(4.2)
def add(n1, n2):
    new = n1 + n2
    return new
add(4, 2)
b = int
b(4.2)
```

```python
# 轉換操作: Series.apply(轉換的SOP)
def func(n):
    return n ** 0.5
s = pd.Series([1, 2, 3])
s.apply(func)
```
