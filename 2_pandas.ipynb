```python
import urllib.request as req
url = "https://github.com/Elwing-Chou/kmu1202/raw/main/danmu.txt"
req.urlretrieve(url, "danmu.txt")
```

```python
# Everything is an object
# 每個東西都是具有型態
# 型態都具有它特定的操作
# 字串型態: s = "hello", 操作: s[0]
# 字典型態: d = {"height":175} 操作: d["height"]
# 還沒執行的SOP型態: int 操作:(4.2)
def add(n1, n2):
    new = n1 + n2
    return new
add(4, 2)
b = int
b(4.2)
```

```python
# 轉換操作: Series.apply(轉換的SOP)
def func(n):
    return n ** 0.5
s = pd.Series([1, 2, 3])
s.apply(func)
```

```python
import jieba
import urllib.request as req
url = "https://github.com/fxsjy/jieba/raw/master/extra_dict/dict.txt.big"
req.urlretrieve(url, "dict.big")
jieba.set_dictionary("dict.big")
```

```python
def func(s):
    return " ".join(jieba.cut(s))
df["text"] = df["text"].apply(func)
df
```

1. 資料視覺化基礎: matplotlib
2. 進階: ....seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(x=df["time"], bins=100, kde=True)
```

1. title
2. axis: 軸
3. label: 軸標籤
4. ticks: 刻度
5. legend: 圖例

```python
import jieba.analyse
textjoin = "\n".join(df["text"])
keywords = jieba.analyse.extract_tags(textjoin, topK=30, withWeight=True)
tfidf = pd.DataFrame(keywords, columns=["word", "tfidf"])
tfidf
```

```python
# Colab 進行matplotlib繪圖時顯示繁體中文
from matplotlib.font_manager import fontManager

url = "https://drive.google.com/uc?id=1eGAsTN1HBpJAkeVM57_C7ccp7hbgSz3_&export=download"
req.urlretrieve(url, "TaipeiSansTCBeta-Regular.ttf")
fontManager.addfont('TaipeiSansTCBeta-Regular.ttf')
plt.rc('font', family='Taipei Sans TC Beta')
```

```python
plt.figure(figsize=[14, 14])
sns.barplot(x=tfidf["word"], y=tfidf["tfidf"])
plt.xticks(rotation=30)
plt.show()
```
